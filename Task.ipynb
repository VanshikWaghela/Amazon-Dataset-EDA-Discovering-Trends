### Breaking down category and sub-category into 2 columns

```python
categ = df[['amazon_category_and_sub_category']]
```

```python
# Create a copy of the original DataFrame to store the extracted content
subcateg = categ.copy()

# Loop through each row and column to extract content after '>'
for row in subcateg.index:
    for col in subcateg.columns:
        cell_value = subcateg.at[row, col]
        if isinstance(cell_value, str) and '>' in cell_value:
            extracted_part = cell_value.split('>', 1)[-1]
            subcateg.at[row, col] = extracted_part
            categ.at[row, col] = cell_value.replace('>' + extracted_part, '')

# Print or access subcateg and modified categ here
```

```python
subcateg
```

```python
categ
```

```python
df['Category'] = categ
df['Sub_Category'] = subcateg
```
